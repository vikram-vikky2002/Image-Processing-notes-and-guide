# CV2 image properties

```python
img = cv2.imread("img2.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray', vmin=0, vmax=255)

cv2.imshow("window",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Changing different color variants.

```python
# making blue -> 0
imgBlue = cv2.imread("img2.png")
imgBlue[:,:,0] = 0

plt.imshow(imgBlue, vmin=0, vmax=255)
```

```python
# making green -> 0
imgGreen = cv2.imread("img2.png")
imgGreen[:,:,1] = 0

plt.imshow(imgGreen, vmin=0, vmax=255)
```

```python
# making red -> 0
imgRed = cv2.imread("img2.png")
imgRed[:,:,2] = 0

plt.imshow(imgRed, vmin=0, vmax=255)
```

```python
combined_img = np.hstack((imgBlue, imgGreen, imgRed))

plt.imshow(combined_img, vmin=0, vmax=255)
cv2.imshow("window", combined_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Resize image

```python
img_resize = cv2.resize(color_img, (512, 512))
img_resize = cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB)

plt.imshow(img_resize, vmin=0, vmax=255)

cv2.imshow("window", img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### resize by percentage

```python
img_resize = cv2.resize(color_img, (color_img.shape[1]//2, color_img.shape[0]//2))
img_resize = cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB)

plt.imshow(img_resize, vmin=0, vmax=255)

cv2.imshow("window", img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Crop Image

```python
img = cv2.imread("img2.png")
img_crop = img[30:100, 40:110]
img_crop = cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB)

plt.imshow(img_crop, vmin=0, vmax=255)

cv2.imshow("window", img_crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Save an image

```python
cv2.imwrite("cropped_img.png", img_crop)
```
